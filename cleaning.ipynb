```python
import pandas as pd

file_path ="stock_and_econ.xlsx"

df= pd.read_excel(file_path)

print(df.head())
```

```python
print(df.columns)
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
print(missing_percentage.sort_values(ascending=False))
```

```python
#drop columns with more than 50% missing values
df=df.dropna(thresh=len(df)*0.5, axis=1)
```

```python
print(df.columns)
```

```python
#impute missing values with mean for numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
#impute missing values with mode for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])
print(df.columns)
```

```python
#drop irrelevant columns
```

```python
# Select only numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Print the correlation matrix
print(correlation_matrix)
```

```python
high_corr_pairs = correlation_matrix[correlation_matrix > 0.9]
print(high_corr_pairs)
```

```python
# Drop rows and columns with all NaN values in the correlation matrix
valid_correlation_matrix = correlation_matrix.dropna(how='all', axis=0).dropna(how='all', axis=1)

# Print the cleaned correlation matrix
print(valid_correlation_matrix)
```

```python
selected_columns = [
    'Market Value (%)', 'P/E', 'P/B', 'ROE', 'ROA', 'Profit Margin',
    'Debt/Equity', 'FCF Gr', 'EBITDA Gr', 'Net Inc Gr', 'Op Inc Gr',
    'GDP CYOY Index', 'CPI YOY Index', 'PPI YOY Index', 'Div Yld',
    'Div Pay Ratio', 'Free CF', 'Year', 'Quarter'
]
```

```python
final_df = df[selected_columns]
```

```python
final_df.to_csv("final_dataset_v1.csv", index=False)
```
